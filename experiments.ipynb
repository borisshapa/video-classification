```python
import torch

from models.linear_classifier import LinearClassifier
from models.transformer_encoder import TransformerEncoder
from video_classification_utils.train import train_on_embeddings_dataset, train_on_images_dataset
```

```python
num_classes = 4
batch_size = 16
epochs = 10
alpha = 0.4
log_every = 20
device = "cuda" if torch.cuda.is_available() else "cpu"
```

## Linear + embeddings mixup

```python
linear_model = LinearClassifier(512, num_classes).to(device)
train_on_embeddings_dataset(
    model=linear_model,
    average_embeddings=True,
    mixup=True,
    batch_size = batch_size,
    epochs = epochs,
    alpha = alpha,
    device = device,
    log_every = log_every
)
```

## Linear w/o mixup

```python
linear_model = LinearClassifier(512, num_classes).to(device)
train_on_embeddings_dataset(
    model=linear_model,
    average_embeddings=True,
    mixup=False,
    batch_size = batch_size,
    epochs = epochs,
    alpha = alpha,
    device = device,
    log_every = log_every
)
```

## Transformer Encoder + embedding mixup

```python
transformer_model = TransformerEncoder(
    transformer_layers = 1,
    emb_size = 512,
    max_len = 100,
    num_classes = 4
).to(device)
train_on_embeddings_dataset(
    model=transformer_model,
    average_embeddings=False,
    mixup=True,
    batch_size = batch_size,
    epochs = epochs,
    alpha = alpha,
    device = device,
    log_every = log_every
)
```

## Transformer Encoder w/o mixup

```python
transformer_model = TransformerEncoder(
    transformer_layers = 1,
    emb_size = 512,
    max_len = 100,
    num_classes = 4
).to(device)
train_on_embeddings_dataset(
    model=transformer_model,
    average_embeddings=False,
    mixup=False,
    batch_size = batch_size,
    epochs = epochs,
    alpha = alpha,
    device = device,
    log_every = log_every
)
```

## RANDOM FRAME Linear + mixup 

```python
linear_model = LinearClassifier(512, num_classes).to(device)
train_on_images_dataset(
    model=linear_model,
    mixup=True,
    batch_size = batch_size,
    epochs = epochs,
    alpha = alpha,
    device = device,
    log_every = log_every
)
```

## RANDOM FRAME Linear w/o mixup

```python
linear_model = LinearClassifier(512, num_classes).to(device)
train_on_images_dataset(
    model=linear_model,
    mixup=False,
    batch_size = batch_size,
    epochs = epochs,
    alpha = alpha,
    device = device,
    log_every = log_every
)
```
